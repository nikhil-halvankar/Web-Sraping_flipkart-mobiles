```python
# import important lib.

from urllib.request import urlopen
from bs4 import BeautifulSoup as soup
```

```python
# reading url

httpdata = urlopen("https://www.flipkart.com/search?q=mobiles&as=on&as-show=on&otracker=AS_Query_TrendingAutoSuggest_5_0_na_na_na&otracker1=AS_Query_TrendingAutoSuggest_5_0_na_na_na&as-pos=5&as-type=TRENDING&suggestionId=mobiles&requestId=a6f4b5f4-9990-4fd1-b1ae-b0b97f3be0bb&p%5B%5D=facets.price_range.from%3DMin&p%5B%5D=facets.price_range.to%3D10000")
webdata = httpdata.read()
print(webdata)
```

```python
soup1 = soup(webdata)   # convert into soup
```

```python
# getting all page link data
pages_link = soup1.findAll('a',{'class':'ge-49M'})
domain = 'https://www.flipkart.com/search?q=mobiles&as=on&as-show=on&otracker=AS_Query_TrendingAutoSuggest_5_0_na_na_na&otracker1=AS_Query_TrendingAutoSuggest_5_0_na_na_na&as-pos=5&as-type=TRENDING&suggestionId=mobiles&requestId=a6f4b5f4-9990-4fd1-b1ae-b0b97f3be0bb&p%5B%5D=facets.price_range.from%3DMin&p%5B%5D=facets.price_range.to%3D10000&page='

for i in range(2,42):
    link = domain + str(i)
    page_data = urlopen(link)
    webdata1 = page_data.read()
    webdata += webdata1
```

```python
soupdata = soup(webdata,'html.parser')
type(soupdata)
```

```python
containers = soupdata.findAll('div',{'class':'_2kHMtA'})
print(type(containers),len(containers))          # check if got all  mobiles of pages 
```

```python
containers[0]             # data of 1st mobile
```

```python
f = open('mobiles_under_10000.csv','wb')
f.write('productName,star,Ratings,Reviews,price,originalmrp,disc,ram,display,camera,battery,processor\n'.encode())

for container in containers:
    # finding product name
    product = container.findAll('div',{'class':'_4rR01T'}) # findAll extract all info within class in list type
    productName = product[0].text.replace(',','')
    
    # finding stars 
    stars = container.find('div',{'class':'_3LWZlK'})  # find extract only 1st occurance ie tag.
    try:
        star = stars.text
    except:
        star = 0
    
    # finding ratings and reviews
    rating = container.find('span',{'class':'_2_R_DZ'})
    try:
        ratrev = re.findall('\d+,*\d*,*\d*',rating.text)   # see below for better understand
        Ratings = ratrev[0].replace(',','')
        Reviews = ratrev[1].replace(',','')
    except:
        Ratings = 0
        Reviews = 0
    
    # finding price
    price = container.find('div',{'class':'_30jeq3 _1_WHN1'}).text.replace(',','')
    
    # finding mrp
    mrp = container.find('div',{'class':'_3I9_wc _27UcVY'})
    try:
        originalmrp = mrp.text.replace(',','')
    except:
        originalmrp = 0
       
    # finding dicount
    discount = container.find('div',{'class':'_3Ay6Sb'})
    try:
        disc = discount.text
    except:
        disc = 0
        
    # finding info
    info = container.findAll('li',{'class':'rgWa7D'})
    try:
        ram = info[0].text
        display = info[1].text
        camera = info[2].text
        battery = info[3].text
        processor = info[4].text
        
    except:
        ram = 0
        display = 0
        camera = 0
        battery = 0
        processor = 0
        
    print(productName,star,Ratings,Reviews,price,originalmrp,disc,ram,display,camera,battery,processor)
    f.write(f"{productName},{star},{Ratings},{Reviews},{price},{originalmrp},{disc},{ram},{display},{camera},{battery},{processor}\n".encode())
    print("\n")
    
f.close()
```
